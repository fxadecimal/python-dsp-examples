```python
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (6, 6)

import sys, os
from numpy import *
import matplotlib.pyplot as plt
from scipy.signal import tf2zpk #butter, freqz, freqs, 

plt.grid()


def zplane(*args):
    circle1 = plt.Circle((0, 0), 1, color='b', fill=False)
    plt.gcf().gca().add_artist(circle1)
    
    z,p,k = tf2zpk(*args)
    plt.scatter(real(p),imag(p),marker='o')
    plt.scatter(real(z),imag(z),marker='x')

# setup plot
plt.axis('equal')
plt.axes().set_ylim([-1,1])
plt.axes().set_xlim([-1,1])

# create some coeffs
w = pi/4
b = array([1.0, -cos(w)])
a = array([1, -2*cos(w), 1.0])

zplane(b,a)
```
